```python
import pdfplumber
```

```python
# Open the PDF file
pdf_file = "/Users/aditkotwal/Projects/FinancialAnalysisApp/aapl_report.pdf"  # Path to your PDF file
with pdfplumber.open(pdf_file) as pdf:
    # Extract text from all pages
    text = ""
    for page in pdf.pages:
        text += page.extract_text()
```

```python
text
```

```python
# Search for the "business risks" section
search_term = "business risks"
start_index = text.lower().find(search_term.lower())

if start_index != -1:
    # Extract a portion of the text where "business risks" is found
    section_text = text[start_index:start_index + 500]  # Adjust length as needed
    print(f"Extracted text from 'business risks' section:\n{section_text}")
else:
    print(f"'{search_term}' not found in the document.")
```
